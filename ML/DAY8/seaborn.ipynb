LINE PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data={
    'Name':['Alice','Bob','Charlie','David'],
    'Age':[25,30,33,47]
}
df=pd.DataFrame(data)
sns.lineplot(x=df.index,y='Age',data=df)
plt.title('Age of Individual')
plt.xlabel('Index')
plt.ylabel('Age')
plt.show()
```

KERNAL DENSITY ESTIMATION

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load iris dataset
iris = load_iris()

# Create dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add species column
df['species'] = pd.Series(iris.target).map({
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
})

# KDE plot
sns.kdeplot(
    data=df[df['species'] == 'virginica']['sepal length (cm)'],
    label='virginica',
    shade=True
)

plt.legend()
plt.title('KDE plot of sepal length for virginica')
plt.xlabel('sepal length (cm)')
plt.ylabel('Density')
plt.show()
```

CUSTOMISE THE SEABORN

Adding title and axis labels

```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load iris dataset
iris = sns.load_dataset('iris')
sns.scatterplot(x='species',y='sepal_length',data=iris,hue='species')
plt.title('scatter plot of sepal length for virginica')
plt.xlabel('sepal length (cm)')
plt.ylabel('Density')
plt.show()
```
